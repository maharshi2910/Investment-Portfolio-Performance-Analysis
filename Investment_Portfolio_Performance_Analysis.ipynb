## Objective

The objective of this project is to analyze an investment portfolio using Python and Pandas. The project calculates profit/loss, return percentage, identifies the best and worst performing investments, and visualizes portfolio performance using charts to support financial decision-making.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Asset": [
        "Reliance",
        "TCS",
        "Gold ETF",
        "Forex Account",
        "Bitcoin",
        "SBI Mutual Fund",
        "Silver ETF",
        "HDFC Bank"
    ],

    "Category": [
        "Stock",
        "Stock",
        "Gold",
        "Forex",
        "Crypto",
        "Mutual Fund",
        "Commodity",
        "Stock"
    ],

    "Investment":[50000,40000,30000,60000,20000,25000,15000,35000],

    "Current Value":[58000,43000,34500,69000,18000,28000,17000,39000]
}

df = pd.DataFrame(data)

df
```

```python
df.to_csv("investment_portfolio.csv", index=False)

print("Portfolio dataset saved successfully!")
```

```python
df = pd.read_csv("investment_portfolio.csv")

df.head()
```

```python
# Display basic information about the dataset

print("First 5 Records:")
print(df.head())

print("\nDataset Information:")
df.info()

print("\nStatistical Summary:")
print(df.describe())
```

```python
# Calculate Profit/Loss

df["Profit/Loss"] = df["Current Value"] - df["Investment"]

# Calculate Return Percentage

df["Return %"] = (df["Profit/Loss"] / df["Investment"]) * 100

df
```

```python
best = df.loc[df["Return %"].idxmax()]
worst = df.loc[df["Return %"].idxmin()]

print("Best Performing Investment")
print(best)

print("\nWorst Performing Investment")
print(worst)
```

```python
plt.figure(figsize=(10,5))

plt.bar(df["Asset"], df["Profit/Loss"])

plt.title("Profit/Loss of Investments")
plt.xlabel("Assets")
plt.ylabel("Profit/Loss (₹)")

plt.xticks(rotation=30)

plt.show()
```

```python
plt.figure(figsize=(7,7))

plt.pie(
    df["Investment"],
    labels=df["Asset"],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Investment Distribution")

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.bar(df["Asset"], df["Return %"])

plt.title("Return Percentage of Investments")
plt.xlabel("Assets")
plt.ylabel("Return %")

plt.xticks(rotation=30)

plt.show()
```

```python
category = df.groupby("Category")["Investment"].sum()

print(category)
```

```python
category.plot(kind="bar", figsize=(8,5))

plt.title("Category Wise Investment")

plt.xlabel("Category")

plt.ylabel("Investment (₹)")

plt.show()
```

```python
print("Total Investment :", df["Investment"].sum())

print("Current Portfolio Value :", df["Current Value"].sum())

print("Total Profit :", df["Profit/Loss"].sum())

print("Average Return :", round(df["Return %"].mean(),2), "%")
```
